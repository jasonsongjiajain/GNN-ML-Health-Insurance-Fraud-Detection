```python
inpatient_columns = set(train_inpatient.columns)
outpatient_columns = set(train_outpatient.columns)
common_columns = inpatient_columns.intersection(outpatient_columns)
inpatient_only = inpatient_columns - outpatient_columns
outpatient_only = outpatient_columns - inpatient_columns

print(f"Common columns: {len(common_columns)}")
print(f"Columns only in inpatient: {inpatient_only}")
print(f"Columns only in outpatient: {outpatient_only}")
```

```python
# For inpatient data, add any columns that are in outpatient but not in inpatient
for col in outpatient_only:
    train_inpatient[col] = None
    
# For outpatient data, add any columns that are in inpatient but not in outpatient
for col in inpatient_only:
    train_outpatient[col] = None

train_inpatient['IP_indicator'] = 1  # 1 for inpatient
train_outpatient['IP_indicator'] = 0  # 0 for outpatient
```

```python
# Concatenate the two dataframes
combined_claims = pd.concat([train_inpatient, train_outpatient], ignore_index=True)

# Check the shape of the combined dataset
print(f"Combined dataset shape: {combined_claims.shape}")
```

```python
# # Step 1: Identify the key beneficiary columns to include
# beneficiary_cols = ['BeneID', 'DOB', 'Gender', 'Race', 'State', 'County', 
#                    'ChronicCond_Alzheimer', 'ChronicCond_Heartfailure', 
#                    'ChronicCond_KidneyDisease', 'ChronicCond_Cancer',
#                    'ChronicCond_ObstrPulmonary', 'ChronicCond_Depression',
#                    'ChronicCond_Diabetes', 'ChronicCond_IschemicHeart',
#                    'ChronicCond_Osteoporasis', 'ChronicCond_rheumatoidarthritis',
#                    'ChronicCond_stroke']

# Step 2: Merge the combined claims data with beneficiary data
combined_with_beneficiary = combined_claims.merge(
    train_beneficiary_data, 
    on='BeneID',  # Join on the beneficiary ID
    how='left'    # Use left join to keep all claims records
)

# Check the shape of the merged dataset
print(f"Combined with beneficiary dataset shape: {combined_with_beneficiary.shape}")
```

```python
# Step 3: Merge with Provider fraud labels
final_data = combined_with_beneficiary.merge(
    train_provider_label,  # Dataset containing provider fraud labels
    on='Provider',         # Join on the Provider ID
    how='left'             # Use left join to keep all claims records
)

# Check the shape and preview the final dataset
print(f"Final dataset shape: {final_data.shape}")
print("\nSample of final dataset:")
final_data.head()

# Check the distribution of fraud labels
fraud_count = final_data['PotentialFraud'].value_counts()
fraud_distribution = final_data['PotentialFraud'].value_counts(normalize=True) * 100
print(f"\nFraud distribution (%):\n{fraud_count}")
print(f"\nFraud distribution (%):\n{fraud_distribution}")
```

```python
# Change Physician ID to INT
final_data['AttendingPhysician'] = final_data['AttendingPhysician'].str.removeprefix("PHY").fillna(0).astype(int) + 3_000_000
final_data['OperatingPhysician'] = final_data['OperatingPhysician'].str.removeprefix("PHY").fillna(0).astype(int) + 3_000_000
final_data['OtherPhysician'] = final_data['OtherPhysician'].str.removeprefix("PHY").fillna(0).astype(int) + 3_000_000

# Change Beneficiary ID to INT
final_data['BeneID'] = final_data['BeneID'].str.removeprefix("BENE").fillna(0).astype(int) +2_000_000

# Change Provider ID to INT
final_data['Provider'] = final_data['Provider'].str.removeprefix("PRV").astype(int) + 1_000_000

# Change Beneficiary ID to INT
final_data['ClaimID'] = final_data['ClaimID'].str.removeprefix("CLM").fillna(0).astype(int) +4_000_000
```

```python
final_data['AttendingPhysician'] = final_data['AttendingPhysician'].fillna(0)
final_data['OperatingPhysician'] = final_data['OperatingPhysician'].fillna(0)
final_data['OtherPhysician'] = final_data['OtherPhysician'].fillna(0)

for i in range(1, 11):
    col = f'ClmDiagnosisCode_{i}'
    if col in final_data.columns:
        final_data[col] = final_data[col].fillna('UNKNOWN')
for i in range(1, 7):
    col = f'ClmProcedureCode_{i}'
    if col in final_data.columns:
        final_data[col] = final_data[col].fillna(0)

# Handle ClmAdmitDiagnosisCode
if 'ClmAdmitDiagnosisCode' in final_data.columns:
    final_data['ClmAdmitDiagnosisCode'] = final_data['ClmAdmitDiagnosisCode'].fillna('UNKNOWN')
        
# Fill missing DeductibleAmtPaid with 0
if 'DeductibleAmtPaid' in final_data.columns:
    final_data['DeductibleAmtPaid'] = final_data['DeductibleAmtPaid'].fillna(0)

final_data['RenalDiseaseIndicator'] = final_data['RenalDiseaseIndicator'].map({'Y':1, '0':0})

final_data['Patient_Deceased'] = final_data['DOD'].notna().astype(int)

final_data['DOB'] = pd.to_datetime(final_data['DOB'])
final_data['DOD'] = pd.to_datetime(final_data['DOD'])

# Step 1: Convert dates and compute AdmitForDays
final_data['AdmissionDt'] = pd.to_datetime(final_data['AdmissionDt'], errors='coerce')
final_data['DischargeDt'] = pd.to_datetime(final_data['DischargeDt'], errors='coerce')
final_data['AdmitForDays'] = (final_data['DischargeDt'] - final_data['AdmissionDt']).dt.days

final_data['AdmitForDays'] = final_data['AdmitForDays'].fillna(0)

final_data = final_data.drop(columns=['AdmissionDt','DischargeDt'])

final_data['DiagnosisGroupCode'] = final_data['DiagnosisGroupCode'].fillna('UNKNOWN')

final_data = final_data.drop(columns=['DOD'])
```

```python
final_data.isnull().sum()
```

```python
# assume final_data is already cleaned, with no nulls
from sklearn.model_selection import train_test_split


providers = final_data['Provider'].unique()
train_p, rest_p = train_test_split(providers, test_size=0.4, random_state=42)
val_p,  test_p = train_test_split(rest_p,    test_size=0.5, random_state=42)

final_data['split'] = final_data['Provider'].map(
    lambda p: 'train' if p in train_p
              else ('val'  if p in val_p else 'test')
)

df_train = final_data[final_data.split=='train'].drop(columns='split')
df_val   = final_data[final_data.split=='val'].drop(columns='split')
df_test  = final_data[final_data.split=='test'].drop(columns='split')
```

### Without ID's

```python
# These are the columns we never feed into the model:
id_cols = ['Provider','BeneID','ClaimID','AttendingPhysician', 'OperatingPhysician', 'OtherPhysician' ]

# Separate X/y for each split:
X_train = df_train.drop(columns=id_cols + ['PotentialFraud'] + ['ClaimStartDt'] + ['ClaimEndDt'])
y_train = df_train['PotentialFraud']

X_val   = df_val.drop(columns=id_cols + ['PotentialFraud'] + ['ClaimStartDt'] + ['ClaimEndDt'])
y_val   = df_val['PotentialFraud']

X_test  = df_test.drop(columns=id_cols + ['PotentialFraud'] + ['ClaimStartDt'] + ['ClaimEndDt'])
y_test  = df_test['PotentialFraud']
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from scipy import sparse

# CATEGORICAL features to one-hot (strings or low-cardinality)
cat_cols = [
    'ClmAdmitDiagnosisCode'
    'DiagnosisGroupCode',
    'ClmDiagnosisCode_1',
    'ClmDiagnosisCode_2',
    'ClmDiagnosisCode_3',
    'ClmDiagnosisCode_4',
    'ClmDiagnosisCode_5',
    'ClmDiagnosisCode_6',
    'ClmDiagnosisCode_7',
    'ClmDiagnosisCode_8',
    'ClmDiagnosisCode_9',
    'ClmDiagnosisCode_10',
]

# NUMERICAL features to scale
num_cols = [
    'InscClaimAmtReimbursed',
    'AdmitForDays',
    'DeductibleAmtPaid',
    'ClmProcedureCode_1',
    'ClmProcedureCode_2',
    'ClmProcedureCode_3',
    'ClmProcedureCode_4',
    'ClmProcedureCode_5',
    'ClmProcedureCode_6',
    'IP_indicator',
    'Gender',
    'Race',
    'State',
    'County'
    'AdmitForDays',
    'NoOfMonths_PartACov',
    'NoOfMonths_PartBCov',
    'RenalDiseaseIndicator',
    'ChronicCond_Alzheimer',
    'ChronicCond_Heartfailure',
    'ChronicCond_KidneyDisease',
    'ChronicCond_Cancer',
    'ChronicCond_ObstrPulmonary',
    'ChronicCond_Depression',
    'ChronicCond_Diabetes',
    'ChronicCond_IschemicHeart',
    'ChronicCond_Osteoporasis',
    'ChronicCond_rheumatoidarthritis',
    'ChronicCond_stroke',
    'IPAnnualReimbursementAmt',
    'IPAnnualDeductibleAmt',
    'OPAnnualReimbursementAmt',
    'OPAnnualDeductibleAmt',
    'Patient_Deceased'
]


# 1) One-hot encode your cats (as you already do)
ohe = OneHotEncoder(handle_unknown='ignore', sparse_output=True)
ohe.fit(X_train[cat_cols])
X_train_cat = ohe.transform(X_train[cat_cols])
X_val_cat   = ohe.transform(X_val[cat_cols])
X_test_cat  = ohe.transform(X_test[cat_cols])

# 2) Scale your numerics on the raw DataFrame
scaler = StandardScaler()
X_train_num_scaled = scaler.fit_transform(X_train[num_cols])
X_val_num_scaled   = scaler.transform(X_val[num_cols])
X_test_num_scaled  = scaler.transform(X_test[num_cols])

# 3) Convert scaled arrays to sparse
X_train_num_sp = sparse.csr_matrix(X_train_num_scaled)
X_val_num_sp   = sparse.csr_matrix(X_val_num_scaled)
X_test_num_sp  = sparse.csr_matrix(X_test_num_scaled)

# 4) Finally stack
X_train_enc = sparse.hstack([X_train_num_sp, X_train_cat ], format='csr')
X_val_enc   = sparse.hstack([X_val_num_sp,       X_val_cat],   format='csr')
X_test_enc  = sparse.hstack([X_test_num_sp,  X_test_cat],  format='csr')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, roc_auc_score

clf = LogisticRegression()  # defaults: no class_weight, threshold=0.5
clf.fit(X_train_enc, y_train)

# Evaluate
for X, y, name in [(X_val_enc, y_val, 'VAL'), (X_test_enc, y_test, 'TEST')]:
    preds = clf.predict(X)
    probs = clf.predict_proba(X)[:,1]
    print(f"--- {name} ---")
    print(classification_report(y, preds))
    print("AUROC:", roc_auc_score(y, probs))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import classification_report, roc_auc_score

# 1) Instantiate the GBM (you can tune n_estimators, max_depth, etc. if desired)
gbm = GradientBoostingClassifier(
    n_estimators=100,       # number of trees
    learning_rate=0.1,      # shrinkage
    max_depth=3,            # tree depth
    random_state=42
)

# 2) Fit on the same encoded training data
gbm.fit(X_train_enc, y_train)

# 3) Evaluate on validation and test splits
for X, y, name in [(X_val_enc, y_val, 'VAL'), (X_test_enc, y_test, 'TEST')]:
    preds = gbm.predict(X)
    probs = gbm.predict_proba(X)[:, 1]
    print(f"--- {name} ---")
    print(classification_report(y, preds))
    print("AUROC:", roc_auc_score(y, probs))
```

## Feature Selection

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, roc_auc_score

# 1) Feature selection
fs_clf = RandomForestClassifier(n_estimators=100, random_state=42)
sfm = SelectFromModel(fs_clf, threshold='median')
sfm.fit(X_train_enc, y_train)

X_train_sel = sfm.transform(X_train_enc)
X_val_sel   = sfm.transform(X_val_enc)
X_test_sel  = sfm.transform(X_test_enc)

# 2) Re-fit & evaluate
final_clf = RandomForestClassifier(n_estimators=100, random_state=42)
final_clf.fit(X_train_sel, y_train)

for X, y, name in [(X_val_sel, y_val, 'VAL'), (X_test_sel, y_test, 'TEST')]:
    preds = final_clf.predict(X)
    probs = final_clf.predict_proba(X)[:,1]
    print(f"--- {name} (selected features) ---")
    print(classification_report(y, preds))
    print("AUROC:", roc_auc_score(y, probs))
```

```python
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, roc_auc_score


# 2) Fit the selector on train only
selector = SelectKBest(chi2, k=100)
X_train_sel = selector.fit_transform(X_train_enc, y_train)
X_val_sel   = selector.transform(X_val_enc)
X_test_sel  = selector.transform(X_test_enc)

# 4) Now train your final model on the reduced feature set
clf = LogisticRegression(random_state=42)
clf.fit(X_train_sel, y_train)

# 5) Evaluate
for X, y, name in [(X_val_sel, y_val, 'VAL'), (X_test_sel, y_test, 'TEST')]:
    preds = clf.predict(X)
    probs = clf.predict_proba(X)[:,1]
    print(f"--- {name} ---")
    print(classification_report(y, preds))
    print("AUROC:", roc_auc_score(y, probs))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.feature_selection import SelectFromModel
from sklearn.metrics import classification_report, roc_auc_score

# 1) Use a shallow tree to rank feature importances
dt_selector = DecisionTreeClassifier(max_depth=5, random_state=42)
selector = SelectFromModel(estimator=dt_selector, threshold='median')

# fit on the full encoded training matrix
selector.fit(X_train_enc, y_train)

# 2) Transform each split
X_train_sel = selector.transform(X_train_enc)
X_val_sel   = selector.transform(X_val_enc)
X_test_sel  = selector.transform(X_test_enc)

# 3) Train your final Decision Tree on the reduced feature set
dt_final = DecisionTreeClassifier(random_state=42)
dt_final.fit(X_train_sel, y_train)

# 4) Evaluate on validation and test
for X_sel, y, split in [(X_val_sel, y_val, 'VAL'), (X_test_sel, y_test, 'TEST')]:
    preds = dt_final.predict(X_sel)
    probs = dt_final.predict_proba(X_sel)[:, 1]
    print(f"--- {split} ---")
    print(classification_report(y, preds))
    print("AUROC:", roc_auc_score(y, probs))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.feature_selection import SelectFromModel
from sklearn.metrics import classification_report, roc_auc_score

# 1) Build a gradient booster to rank features
gb_selector = GradientBoostingClassifier(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=3,
    random_state=42
)

# Wrap it in SelectFromModel, keeping only features â‰¥ median importance
selector = SelectFromModel(estimator=gb_selector, threshold="median")

# Fit the selector on the full training set
selector.fit(X_train_enc, y_train)

# 2) Reduce all splits to the selected feature subset
X_train_sel = selector.transform(X_train_enc)
X_val_sel   = selector.transform(X_val_enc)
X_test_sel  = selector.transform(X_test_enc)

print("Kept features:", X_train_sel.shape[1], "of", X_train_enc.shape[1])

# 3) Train a fresh GradientBoostingClassifier on the selected features
gb_final = GradientBoostingClassifier(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=3,
    random_state=42
)
gb_final.fit(X_train_sel, y_train)

# 4) Evaluate on validation and test
for X_sel, y, split in [(X_val_sel, y_val, 'VAL'), (X_test_sel, y_test, 'TEST')]:
    preds = gb_final.predict(X_sel)
    probs = gb_final.predict_proba(X_sel)[:, 1]
    print(f"--- {split} ---")
    print(classification_report(y, preds))
    print("AUROC:", roc_auc_score(y, probs))
```
