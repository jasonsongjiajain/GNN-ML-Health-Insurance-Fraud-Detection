# Cleaning Missing Values

```python
# Clean Train Dataset
# 1. Missing Values for DOD
# Less amount of Day of Death means more people are alive
train_beneficiary_data['isAlive'] = train_beneficiary_data['DOD'].isna().astype(int)

# 2. 
# Replace RenalDiseaseIndicator unique 
train_beneficiary_data['RenalDiseaseIndicator'] = train_beneficiary_data['RenalDiseaseIndicator'].map({'Y':1, '0': 0})
```

```python
# Handling Null Values
train_outpatient['AttendingPhysician'] = train_outpatient['AttendingPhysician'].fillna('PHY_UNKNOWN')
train_outpatient['OperatingPhysician'] = train_outpatient['OperatingPhysician'].fillna('PHY_UNKNOWN')
train_outpatient['OtherPhysician'] = train_outpatient['OtherPhysician'].fillna('PHY_UNKNOWN')

# Create binary flags for diagnosis codes
for i in range(2, 11):
    col = f'ClmDiagnosisCode_{i}'
    train_outpatient[f'{col}_Present'] = train_outpatient[col].notna().astype(int)

for i in range(1, 7):
    col = f'ClmProcedureCode_{i}'
    train_outpatient[f'{col}_Present'] = train_outpatient[col].notna().astype(int)

# For ClmAdmitDiagnosisCode, create a presence flag and fill missing with 'UNKNOWN'
train_outpatient['ClmAdmitDiagnosisCode_Present'] = train_outpatient['ClmAdmitDiagnosisCode'].notna().astype(int)
train_outpatient['ClmAdmitDiagnosisCode'] = train_outpatient['ClmAdmitDiagnosisCode'].fillna('UNKNOWN')
```

```python
# Step 3: Clean inpatient data
# Similar approach as outpatient data
train_inpatient['AttendingPhysician'] = train_inpatient['AttendingPhysician'].fillna('PHY_UNKNOWN')
train_inpatient['OperatingPhysician'] = train_inpatient['OperatingPhysician'].fillna('PHY_UNKNOWN')
train_inpatient['OtherPhysician'] = train_inpatient['OtherPhysician'].fillna('PHY_UNKNOWN')

# Handle DeductibleAmtPaid - fill with median
train_inpatient['DeductibleAmtPaid'] = train_inpatient['DeductibleAmtPaid'].fillna(train_inpatient['DeductibleAmtPaid'].median())

# Create binary flags for diagnosis codes 2-10
for i in range(2, 11):
    col = f'ClmDiagnosisCode_{i}'
    train_inpatient[f'{col}_Present'] = train_inpatient[col].notna().astype(int)

# Create binary flags for procedure codes
for i in range(1, 7):
    col = f'ClmProcedureCode_{i}'
    train_inpatient[f'{col}_Present'] = train_inpatient[col].notna().astype(int)
```

# Feature Engineering

```python
# Feature Engineering
train_outpatient['LengthOfStay'] = (pd.to_datetime(train_outpatient['ClaimEndDt']) - pd.to_datetime(train_outpatient['ClaimStartDt'])).dt.days

train_inpatient['LengthOfStay'] = (pd.to_datetime(train_inpatient['ClaimEndDt']) - pd.to_datetime(train_inpatient['ClaimStartDt'])).dt.days

# Calculate age of beneficiary
train_beneficiary_data['Age'] = 2025 - pd.to_datetime(train_beneficiary_data['DOB']).dt.year
```

# Nodes

```python
import torch
from torch_geometric.data import HeteroData

# Initialize Heterogeneous Graph
hetero_data = HeteroData()

# Provider node
providers = train_provider_label['Provider'].unique()
provider_idx_map = {provider : i for i, provider in enumerate(providers)}

# Inpatient provider node features
inpatient_provider_feature = train_inpatient.groupby('Provider').agg({
    'InscClaimAmtReimbursed': ['mean', 'std', 'sum', 'count'],
    'DeductibleAmtPaid': ['mean', 'sum'],
    'LengthOfStay': ['mean', 'max']
}).fillna(0)

inpatient_provider_feature.columns = ['_'.join(col).strip() for col in inpatient_provider_feature.columns.values]

# Outpatient provider node features
outpatient_provider_feature = train_outpatient.groupby('Provider').agg({
    'InscClaimAmtReimbursed': ['mean', 'std', 'sum', 'count'],
    'DeductibleAmtPaid': ['mean', 'sum'],
    'LengthOfStay': ['mean', 'max']
}).fillna(0)

outpatient_provider_feature.columns  = ['Outpatient_' + '_'.join(col).strip() for col in outpatient_provider_feature.columns.values]

# Combine Inpatient and Outpatient
provider_feature = inpatient_provider_feature.merge(
    outpatient_provider_feature, left_index = True, right_index = True, how='outer').fillna(0)

provider_feature_tensor = torch.tensor(
    provider_feature.reindex(providers).fillna(0).values, dtype=torch.float
)

# Set label for Provider node 
hetero_data['provider'].x = provider_feature_tensor

```

```python
provider_labels = train_provider_label.set_index('Provider')
provider_labels = provider_labels.reindex(providers)
# Set target node 
provider_labels['PotentialFraud'] = provider_labels['PotentialFraud'].map({'Yes':1, 'No': 0})
hetero_data['provider'].y = torch.tensor(provider_labels['PotentialFraud'].values, dtype=torch.long)
```

```python
from sklearn.preprocessing import OneHotEncoder
import joblib

beneficiaries = train_beneficiary_data['BeneID'].unique()
beneficiary_idx_map = {bene: i for i, bene in enumerate(beneficiaries)}
    
# Create beneficiary features
beneficiary_features = train_beneficiary_data.set_index('BeneID')

# Convert categorical features to one-hot encoding
cat_features = ['Gender', 'Race', 'State', 'County', 'RenalDiseaseIndicator', 'ChronicCond_Alzheimer', 
                   'ChronicCond_Heartfailure', 'ChronicCond_KidneyDisease', 'ChronicCond_Cancer',
                   'ChronicCond_ObstrPulmonary', 'ChronicCond_Depression', 'ChronicCond_Diabetes',
                   'ChronicCond_IschemicHeart', 'ChronicCond_Osteoporasis', 
                   'ChronicCond_rheumatoidarthritis', 'ChronicCond_stroke', 'isAlive']
    
# Numeric features
num_features = ['NoOfMonths_PartACov', 'NoOfMonths_PartBCov', 'IPAnnualReimbursementAmt',
                   'IPAnnualDeductibleAmt', 'OPAnnualReimbursementAmt', 'OPAnnualDeductibleAmt', 'Age']

encoder = OneHotEncoder(sparse_output=False)
cat_data = beneficiary_features[cat_features].fillna('Unknown')
cat_encoded = encoder.fit_transform(cat_data)
    
# Combine with numeric features
num_data = beneficiary_features[num_features].fillna(0).values
beneficiary_feature_matrix = np.hstack([num_data, cat_encoded])
    
# Add beneficiary nodes to graph
hetero_data['beneficiary'].x = torch.tensor(beneficiary_feature_matrix, dtype=torch.float)
```

```python
# Add physician nodes
# Collect all unique physicians
all_physicians = set()
for df in [train_inpatient, train_outpatient]:
        all_physicians.update(df['AttendingPhysician'].dropna().unique())
        all_physicians.update(df['OperatingPhysician'].dropna().unique())
        all_physicians.update(df['OtherPhysician'].dropna().unique())
    
# Remove the unknown physician if present
if 'PHY_UNKNOWN' in all_physicians:
        all_physicians.remove('PHY_UNKNOWN')
    
physicians = list(all_physicians)
physician_idx_map = {phys: i for i, phys in enumerate(physicians)}
    
# Since we don't have specific features for physicians, we'll use an identity feature
hetero_data['physician'].x = torch.eye(len(physicians), dtype=torch.float)
```

## Implement the Graph Atttention model

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.nn import SAGEConv, HeteroConv, Linear
from torch_geometric.nn import HGTConv, GATConv, GCNConv
from torch_geometric.nn import to_hetero
from sklearn.metrics import classification_report, confusion_matrix, precision_recall_fscore_support, roc_auc_score

class GATModel(torch.nn.Module):
    def __init__(self, hidden_channels, out_channels, num_layers, heads=8):
        super().__init__()
        
        self.convs = nn.ModuleList()
        
        # First layer
        self.convs.append(HeteroConv({
            ('provider', 'treats', 'beneficiary'): GATConv((-1, -1), hidden_channels // heads, heads=heads, add_self_loops=False),
            ('beneficiary', 'treated_by', 'provider'): GATConv((-1, -1), hidden_channels // heads, heads=heads, add_self_loops=False),
            ('provider', 'employs', 'physician'): GATConv((-1, -1), hidden_channels // heads, heads=heads, add_self_loops=False),
            ('physician', 'works_for', 'provider'): GATConv((-1, -1), hidden_channels // heads, heads=heads, add_self_loops=False)
        }))
        
        # Hidden layers
        for _ in range(num_layers - 1):
            self.convs.append(HeteroConv({
                ('provider', 'treats', 'beneficiary'): GATConv((-1, -1), hidden_channels // heads, heads=heads, add_self_loops=False),
                ('beneficiary', 'treated_by', 'provider'): GATConv((-1, -1), hidden_channels // heads, heads=heads, add_self_loops=False),
                ('provider', 'employs', 'physician'): GATConv((-1, -1), hidden_channels // heads, heads=heads, add_self_loops=False),
                ('physician', 'works_for', 'provider'): GATConv((-1, -1), hidden_channels // heads, heads=heads, add_self_loops=False)
            }))
        
        # Output layer for binary classification
        self.lin = Linear(hidden_channels, out_channels)
    
    def forward(self, x_dict, edge_index_dict):
        # Apply graph convolutions
        for conv in self.convs:
            x_dict = conv(x_dict, edge_index_dict)
            # Apply ReLU non-linearity
            x_dict = {key: F.relu(x) for key, x in x_dict.items()}
        
        # Only make predictions for providers
        return self.lin(x_dict['provider'])
```

```python
# Initialize model
model = GATModel(hidden_channels=64, out_channels=2, num_layers=2)

# Split data
num_providers = hetero_data['provider'].x.size(0)
perm = torch.randperm(num_providers)
train_idx = perm[:int(0.7 * num_providers)]
val_idx = perm[int(0.7 * num_providers):int(0.85 * num_providers)]
test_idx = perm[int(0.85 * num_providers):]

# Training setup
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
weights = torch.tensor([1.0, 10.0])  # Higher weight for the fraud class
criterion = torch.nn.CrossEntropyLoss(weight=weights)

# Training loop
def train():
    model.train()
    optimizer.zero_grad()
    out = model(hetero_data.x_dict, hetero_data.edge_index_dict)
    loss = criterion(out[train_idx], hetero_data['provider'].y[train_idx])
    loss.backward()
    optimizer.step()
    return float(loss)

# Validation
@torch.no_grad()
def validate():
    model.eval()
    out = model(hetero_data.x_dict, hetero_data.edge_index_dict)
    val_loss = criterion(out[val_idx], hetero_data['provider'].y[val_idx])
    pred = out.argmax(dim=-1)
    val_acc = (pred[val_idx] == hetero_data['provider'].y[val_idx]).sum().item() / val_idx.size(0)
    
    # Get validation metrics for fraud detection
    val_y_true = hetero_data['provider'].y[val_idx].cpu().numpy()
    val_y_pred = pred[val_idx].cpu().numpy()
    val_probs = F.softmax(out, dim=1)[val_idx, 1].cpu().numpy()
    
    precision, recall, f1, _ = precision_recall_fscore_support(
        val_y_true, val_y_pred, average=None
    )
    val_f1_fraud = f1[1]  # F1 for fraud class
    
    return val_loss, val_acc, val_f1_fraud, val_y_true, val_y_pred, val_probs

@torch.no_grad()
def test_model():
    model.eval()
    out = model(hetero_data.x_dict, hetero_data.edge_index_dict)
    test_loss = criterion(out[test_idx], hetero_data['provider'].y[test_idx])
    pred = out.argmax(dim=-1)
    test_acc = (pred[test_idx] == hetero_data['provider'].y[test_idx]).sum().item() / test_idx.size(0)
    
    # Get test metrics for fraud detection
    test_y_true = hetero_data['provider'].y[test_idx].cpu().numpy()
    test_y_pred = pred[test_idx].cpu().numpy()
    test_probs = F.softmax(out, dim=1)[test_idx, 1].cpu().numpy()
    
    # Calculate metrics
    precision, recall, f1, _ = precision_recall_fscore_support(
        test_y_true, test_y_pred, average=None
    )
    test_f1_fraud = f1[1]  # F1 for fraud class
    
    # More detailed evaluation
    class_report = classification_report(
        test_y_true, 
        test_y_pred, 
        target_names=['Non-Fraud', 'Fraud'],
        digits=4
    )
    
    conf_matrix = confusion_matrix(test_y_true, test_y_pred)
    
    # AUC calculation
    try:
        auc = roc_auc_score(test_y_true, test_probs)
    except:
        auc = float('nan')
        
    print("\n--- Test Set Evaluation ---")
    print(f"Test Accuracy: {test_acc:.4f}")
    print(f"Test Loss: {test_loss:.4f}")
    print(f"Test F1 (Fraud): {test_f1_fraud:.4f}")
    print("\nClassification Report:")
    print(class_report)
    print("Confusion Matrix:")
    print(conf_matrix)
    print(f"AUC-ROC: {auc:.4f}")
    
    return test_acc, test_f1_fraud, test_y_true, test_y_pred, test_probs
```

```python
# Training with early stopping
best_val_f1 = 0
patience = 10
counter = 0
train_losses = []
val_losses = []
val_f1_scores = []

for epoch in range(200):
    # Training
    train_loss = train()
    train_losses.append(train_loss)
    
    # Validation
    val_loss, val_acc, val_f1, _, _, _ = validate()
    val_losses.append(val_loss)
    val_f1_scores.append(val_f1)
    
    # Print progress
    if epoch % 5 == 0:
        print(f'Epoch: {epoch:03d}, Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.4f}, Val F1 (Fraud): {val_f1:.4f}')
    
    # Early stopping based on F1 for fraud class
    if val_f1 > best_val_f1:
        best_val_f1 = val_f1
        counter = 0
        # Save the best model
        torch.save(model.state_dict(), 'best_model.pt')
        print(f"Epoch {epoch}: New best model saved with validation F1: {val_f1:.4f}")
        print(f'Epoch: {epoch:03d}, Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.4f}, Val F1 (Fraud): {val_f1:.4f}')
    else:
        counter += 1
        if counter >= patience:
            print(f'Early stopping at epoch {epoch}')
            break

# Load the best model for final evaluation
print("Loading best model for final evaluation...")
model.load_state_dict(torch.load('best_model.pt', weights_only=False))

# Final evaluation on the test set
test_acc, test_f1, test_y_true, test_y_pred, test_probs = test_model()
print(f"Final test accuracy: {test_acc:.4f}")
print(f"Final test F1 score for fraud: {test_f1:.4f}")
```
